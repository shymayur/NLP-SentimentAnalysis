## 2. Importing packages

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sia = SentimentIntensityAnalyzer()
```

## 3. Importing the dataset

```python
df = pd.read_csv('AmazonReviews.csv')
df.head()
```

```python
#sia.polarity_scores('OH!')
```

## 4. Extracting the Polarity Scores of review text

```python
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['Text']
    myid = row['Id']
    res[myid] = sia.polarity_scores(text)
```

```python
res
```

```python
vader = pd.DataFrame(res).T
vader.head()
```

## 5. Merging dataframes as LEFT JOIN under 'Id'

```python
vader = vader.reset_index().rename(columns={'index' : 'Id'})
vader.head()
```

```python
# Now we have sentiment score & the data
vader = vader.merge(df, how='left')
```

```python
vader.head()
```

## 6. Plotting the results of vader (dataframe)

```python
ax = sns.barplot(data=vader, x='Score', y='compound')
ax.set_title('Compound score by Amazon reviews')
plt.show()
```

## 7. Plotting the sentiments with their Scores

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
sns.barplot(data=vader, x='Score', y='pos', ax=axs[0])
sns.barplot(data=vader, x='Score', y='neu', ax=axs[1])
sns.barplot(data=vader, x='Score', y='neg', ax=axs[2])
axs[0].set_title('Positive')
axs[1].set_title('Neutral')
axs[2].set_title('Negative')

plt.tight_layout()
plt.show()
```
